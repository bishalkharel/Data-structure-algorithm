## Squential Data Structures:
### Those data structures which element cannot be access randomly but can be accessed squentially are called squential data structures.

<b> Physical example would be measuring tape because to get into 72 inch you have to go through 1,2,3 inch first</b>

# The Stack



### Some methods of stacks:


<b>The methods are:</b>
<b><ul type="disc">
      <li><ul type="circle">Push(): 
        <li> This method add a new element from last or add a new element so that it becomes a new front facing element</li>
    </ul></li>
    <li><ul type="circle">Pop():
        <li> Remove a last element</li>
        </ul>
    </li>
     <li><ul type="circle">Peek():
        <li>Helps to access the last element or front facing element</li>
        </ul>
    </li>
     <li><ul type="disc">Contains():
        <li> Help us to search element in stack and returns boolean value if our required value is found in stack</li>
        </ul>
    </li>
       
       
       
</ul></b>

## The time complexity for stacks:
<br>
<b>
    1. Accessing
    <ul type="disc"><br>
      <li>If you want to access certain element in stacks then you have to remove element from last so that you can access that particular element. So the time coomplexity will be O(n) because it depends upon our stacks length</li> 
    </ul>
    2. Searching
    <ul type="disc">
      <li>To search the element in stacks the same process goes through so time complexity will be O(n).</li>
    </ul>
    3. Inserting
        <ul type="disc">
            <li> To insert the element you can only do it from last so the time complexity will O(1)</li>
    </ul>
    4. Deleting
        <ul type="disc">
      <li>Same process goes through deleting also because you can only delete from last so it will be O(1)</li>
    </ul>
</b>
