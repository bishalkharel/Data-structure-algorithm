### Methods of queues:
- Enqueue 
- Dequeue
- Peek
- Contains


- <b>Enqueue:
    It adds object in tail which is the entering point of queue
    
- Dequeue: It removes the objects from head which is the ending point of queues. It takes no argument and removes the element at head. In above example if we apply it will remove "now" object.<br><br>
- Peek: It helps us to access the element from head part which is the ending part of queues. It also takes no argument and gives the element at fore most i.e head.<br><br>
- Contains: It searches the element in queues. It return the boolean value if the given object is in queue then it will return True else it will return False
